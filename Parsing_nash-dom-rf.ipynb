```python
import requests
import csv
import pandas as pd
import numpy as np
```

```python
def write_csv(filename, data):
    with open(filename, 'a', encoding='utf-8', newline='') as file:
        csv.DictWriter(file, fieldnames=list(data)).writerow(data)


def main():
    offset = 0
    while True:
        url = 'https://наш.дом.рф/сервисы/api/kn/object'
        params = (
        ('offset',offset),
        ('limit',100),
        ('sortField','devId.devShortCleanNm'),
        ('sortType','asc'),
        ('objStatus', 0)
        
        )
        response = requests.get(url, params=params)
        reestr = response.json()
        list = reestr['data']['list']
        
        for developer in list:
            devId = developer.get('devId','')
            shortName = developer.get('shortName','')
            fullName = developer.get('fullName','')
            groupName = developer.get('groupName','')
            fullForm = developer.get('fullForm','')    
            devInn = developer.get('devInn','')   
            rpdRegionCd = developer.get('rpdRegionCd','') 
            objAddr = developer.get('objAddr','') 
            objFloorMin = developer.get('objFloorMin','') 
            objFloorMax = developer.get('objFloorMax','') 
            objElemLivingCnt = developer.get('objElemLivingCnt','') 
            objReady100PercDt = developer.get('objReady100PercDt','') 
            objSquareLiving = developer.get('objSquareLiving','') 
            latitude = developer.get('latitude','') 
            longitude = developer.get('longitude','') 
            problemFlag = developer.get('problemFlag','') 
            siteStatus = developer.get('siteStatus','') 
            buildType = developer.get('buildType','') 
            
            data = {'devId' : devId, 'shortName' : shortName, 'fullName' : fullName, 'groupName' : groupName,
            'fullForm' : fullForm, 'devInn' : devInn, 'rpdRegionCd' : rpdRegionCd, 'objAddr' : objAddr,
            'objFloorMin' : objFloorMin, 'objFloorMax' : objFloorMax, 'objElemLivingCnt' : objElemLivingCnt,
            'objReady100PercDt' : objReady100PercDt, 'objSquareLiving' : objSquareLiving, 'latitude' : latitude,
            'longitude' : longitude, 'problemFlag' : problemFlag, 'siteStatus' : siteStatus, 'buildType' : buildType }
        
        
            write_csv('nash-dom-rf-s1.csv', data)
        if not list:
            break
            
        offset += 100
    
        
if __name__== '__main__':
    main()
    
```

```python
df = pd.read_csv('nash-dom-rf-s1.csv',header=None)
```
